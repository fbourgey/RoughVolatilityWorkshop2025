### The quadratic rough Heston model


The quadratic rough Heston (QRH) model of <span>[QRH]<sup id="cite_ref-QRH" class="reference"><a href="#cite_note-QRH">[9]</a></sup> may be written as
<p>

$$
\begin{equation}\tag{1}
\begin{aligned}
\frac{dS_t}{S_t} &= -\sqrt{V_t}\,dW_t, \\
V_t &= Y_t ^2 + \uv,
\label{eq:QRHestondef}
\end{aligned}    
\end{equation}
$$

where the minimum variance $ \uv \geq 0$, and
$$
\begin{equation}
Y_t = \bar Y +\int_{-\infty}^t \kappa(t-s)\,\sqrt{V_s}\,dW_s
 =\bar Y -\int_{-\infty}^t \kappa(t-s)\,\frac{dS_s}{S_s}
\end{equation}
$$

is a weighted average of historical returns
and $\kappa(\cdot)$ a kernel function.

### Dynamics in martingale form

Taking the conditional expectation of $Y_u$ for $u > t$ gives,
<p>
\begin{align}
Y_u &=
y_t(u) + \int_{t}^u \kappa(u-s)\,\sqrt{V_s} \, dW_s
\label{eq:Yu}
\end{align}
    <p>
where 
$$
y_t(u) 
:= \eet{Y_u} 
=  \bar Y+\int_{-\infty}^t \kappa(u-s)\,\sqrt{V_s}\,dW_s.
$$
<p>        
Then, $y_\cdot(u)$ is a martingale and 
<p>    
$$
dy_t(u) = \kappa(u-t)\,\sqrt{V_t}\,dW_t .%= \kappa(u-t)\,y_t(t)\,dW_t.
$$

### The variance of $Y_u$

Thus $\eet{Y_u}=y_t(u)$ and from (1) and Itô's isometry,
<p>
$$
\var_t[Y_u] =  \int_t^u\,\xi_t(s)\,\kappa(u-s)^2\,ds,
$$
 <p>   
 where $\xi_t(s) := \eet{V_s}$ for all $s\geq t$.

 - Under QR Heston, we have the mean and variance of $Y_u$ in (quasi-) closed form.

     
  - We will sometimes refer to the curve $y_t(u)$ as the *forward volatility curve*.



### The forward variance curve

From the model definition, for $u>t$, $
V_u = y_u(u)^2+\uv$ so applying Itô's formula, we have
<p>
\begin{align}
\xi_t(u) &= \eet{y_u(u)^2}+\uv \nonumber\\
&= y_t(u)^2 + \eet{\int_t^u\,d\angl{y_{\cdot}(u)}_s}+\uv\nonumber\\
&= y_t(u)^2 +\int_t^u\,\xi_t(s)\,\kappa(u-s)^2\,ds+\uv.
\label{eq:xitu}    
\end{align}
    <p>
Alternatively,
<p>
        $$
 y_t(u)^2 = \xi_t(u)  -  \int_t^u\,\xi_t(s)\,\kappa(u-s)^2\,ds-\uv,
 $$
    
 - Thus $y_t(u)$ may be easily imputed from the forward variance curve.


### The stationary state

In the stationary state,
$
\ee{Y_t} = \bar Y 
$
and 
\begin{align}
\bar V := \ee{V_t} &= \ee{\uv+Y_t^2} \notag\\
&=\uv+\bar Y^2 + \int_{-\infty}^t\,\kappa(t-s)^2\,\ee{V_s}\,ds \notag\\
&=\uv+\bar Y^2 + \|\kappa^2\|\,\bar V.
\label{eq:Vbar}
\end{align}
where
$$
\|\kappa^2\| = \int_0^{\infty}\,\kappa(s)^2\,ds.
$$





### Admissibility condition on the kernel

-  A necessary condition on the kernel then follows:
<p>
\begin{align*}
\|\kappa^2\|  < 1.
\end{align*}
 - Violated by the power-law kernel!


### Pricing with the QRH model

- Like all rough volatility moodels, to compute quantities of interest, we need to define the state vector and the model (kernel) parameters.

    - The (infinite) state vector is the forward variance curve $\xi_t(u)$.

    - The model parameters are the parameters of the chosen kernel.

###  Choice of kernel

- For implementation purposes, for flexibility as well as the existence 
of a closed-form expression for the resolvent kernel $K$, 
we choose 
the gamma kernel given by $$ \kappa(\tau)= \frac{\nu}{\Gamma(\alpha)}\, \tau^{\alpha-1}\,e^{-\lambda \tau}, \tag{2}$$
<p>
with $\alpha = H + \tfrac12$.


- With this choice, the QR Heston model has only four parameters: $H$, $\nu$, $\lambda$, and $\uv$.



### Kernel parameters

- The  QRH model was calibrated to SPX and VIX smiles as of 15-Feb- 2023 by Florian Bourgey of Bloomberg.
    - A non-trivial task.
    - Efficient calibration of the model is work in progress.
    
    
- He found the following QRH parameters (with $c=\uv$):

```python
params0 = {"al": 0.568, "lam": 9.68, "nu": 0.572, "c": 8.1e-3}
```

### Admissibility condition for the gamma kernel

- The admissibilty condition  then evaluates to
$$
\|\kappa^2\|=\int_0^\infty\,\kappa(s)^2\,ds = \cL[\kappa^2](0) = \frac{ \hat \nu^2}{ (2 \lambda)^{2 H}}  <1,
$$
where $\hat{\nu} := \nu \sqrt{ \Gamma (2H)}/\Gamma (\alpha)$.

### With our fitted parameters

We check the necessary condition with our parameters.


```python
def kernel(x, nu, al, lam):
    return (nu / special.gamma(al)) * x ** (al - 1) * np.exp(-lam * x)


def converge_test(params, opt="closed-form"):
    al = params["al"]
    H = al - 1 / 2
    lam = params["lam"]
    nu = params["nu"]

    if opt == "closed-form":
        integral = (
            (nu / special.gamma(al)) ** 2 * special.gamma(2 * H) / (2 * lam) ** (2 * H)
        )
    elif opt == "quad":
        integral = integrate.quad(lambda x: kernel(x, nu, al, lam) ** 2, 0, np.inf)[0]
    else:
        raise ValueError("Check value for opt.")

    assert integral < 1
    return integral


print(converge_test(params0, opt="closed-form"))
print(converge_test(params0, opt="quad"))
```

So our paraameter set is admissible.

### $\xi_t(u)$ from $y_t(u)$

- Recall that

$$
\xi_t(u) = y_t(u)^2 +\int_t^u\,\xi_t(s)\,\kappa(u-s)^2\,ds+\uv. 
$$

- This is a Wiener-Hopf equation for $\xi_t(u)$ whose solution may be written as

$$
\xi_t(u) = y_t(u)^2+\uv+ \int_{t}^u\,K(u-s)\,\left[y_t(s)^2+\uv\right]\,ds
$$

where $K$ is the *resolvent kernel* of $\kappa^2$.    

### Solving the Wiener-Hopf integral equation

- We write the original equation in convolution notation as:
\begin{align*}
\xi
&= y^2 +(\xi \star \kappa^2)+\uv. 
\end{align*}

- The Laplace transform of a convolution is the product of the Laplace transforms.  Thus

\begin{align*}
\cL[\xi]
&= \cL[y^2+\uv] +\cL[\xi]\,\cL[\kappa^2]. 
\end{align*}

- Rearranging gives

\begin{align*}
\cL[\xi]\,\left(1-\cL[\kappa^2]\right)
&= \cL[y^2+\uv] . 
\end{align*}


- So,
<p><p>
\begin{align*}
\cL[\xi] = 
&= \cL[y^2+\uv]+\cL[K]\,\cL[y^2+\uv], 
\end{align*}
<p><p>
where the Laplace transform of the resolvent kernel $K$ is given by
<p>
$$
\cL[K] = \frac{\cL[\kappa^2]}{1-\cL[\kappa^2]}.
$$
    
- Inverting the Laplace transforms gives
    <p>
$$
\xi_t(u) = y_t(u)^2 + \uv + \int_{t}^u\,K(u-s)\,\left[y_t(s)^2+\uv\right]\,ds.
$$
<p>
as required.

### The resolvent of the gamma kernel squared

- The Laplace transform of $\kappa^2$ is given by
$$
   \cL[\kappa^2](p) 
 = \frac{\eta^2\,\Gamma (2 H)}{ (2 \lambda +p)^{2 H}}
    =: \frac{\hat \nu^2}{ (2 \lambda + p)^{2 H}},
$$



- Thus
$$
    \cL[K](p) = \frac{\cL[\kappa^2](p)}{1-\cL[\kappa^2](p)} 
    =\frac{ \hat\nu^2}{(2 \lambda + p)^{2 H }- \hat\nu^2}.
$$

 - This turns out to be the Laplace transform of 
$$
K(\tau) = \hat \nu^2\,e^{-2 \lambda \tau}\,
\tau^{2 H-1}\,E_{2 H,2 H}(\hat\nu^2\,\tau^{2 H}),
$$
<p>
where $E_{\bullet,\bullet}$ is the two-parameter Mittag-Leffler 
function.
    
    
- Thus for the gamma kernel, we have a closed-form expression for the resolvent kernel $K$.

### Dynamics of forward variance

- Recall that
<p>
$$
\xi_t(u) = y_t(u)^2+\uv+ \int_{t}^u\,K(u-s)\,\left[y_t(s)^2+\uv\right]\,ds.
$$

    
    
- Applying Itô's formula and using that $\xi_\cdot(u)$ is a martingale, 
<p>
\begin{align}
d\xi_t(u) &= 2\,y_t(u)\,dy_t(u) + 2\,\int_t^u\,y_t(s)\,dy_t(s)\, K(u-s)\,ds \nonumber\\
&= 2\,\left\{y_t(u)\,\kappa(u-t) + \int_t^u\,y_t(s)\,\kappa(s-t)\,K(u-s)\,ds\right\}\,\sqrt{V_t}\,dW_t \nonumber\\
&= -2\,\left\{\kappa(u-t)\,y_t(u)+\int_t^u\,\kappa(s-t)\, K(u-s)\, y_t(s)\, ds\right\}\,\frac{dS_{t}}{S_{t}} .
\end{align}
<p>

- If $y_t(u)>0$, $dS_{t}$ and $d\xi_t(u)$ are perfectly anti-correlated in the QR Heston model.
    
   - However, sometimes $y_t(u)$ is negative for short-dated $u$.
    
   - This implies in particular that shorter dated smiles should be slighly more symmetric, as observed for example in <span>[Rømer]<sup id="cite_ref-Romer" class="reference"><a href="#cite_note-Romer">[14]</a></sup>.


### A microstructural foundation for the QRH model

- We now show how the QRH model can be obtained as the diffusion limit of a nnonlinear Hawkes process.
    - Just as rough Heston can be obtained as the diffusion limit of a linear Hawkes process.
    - Or more generally, AFV models can be obtained as diffusion limits of AFI models.

### A Hawkes process

We adopt the same Hawkes model of order flow as in Lecture 3 (<span>[AFV]<sup id="cite_ref-AFV" class="reference"><a href="#cite_note-AFV">[10]</a></sup>) ,  but make it quadratic.


- $J_t^{+}$ (buys) and $J_t^{-}$ (sells) are marked jump processes with (possibly different) mark laws $\zeta_+$ and $\zeta_-$.  


- Just as in the AFI case, the log-price process is given in terms of the $J^\pm$ as
<p>
$$
dX_t = -\lambda_{t} m_X dt + dJ_t^+ - dJ_t^-,
$$
    <p>
where $m_X$ is determined by the martingale condition on $S=e^X$. 

### Modeling $V_t$

- In order to complete the definition of the Hawkes order flow model, we need to specify the dynamics of $\lambda_t$ or equivalently $V_t$.


-  We would like a jump model whose diffusion limit is the quadratic rough Heston model (1).


### The quadratic Hawkes order flow (QHO) model 


- Motivated by the form of the QRH $V$  process, we write


\begin{equation}
V_t= v_J\,\lambda_t \;=c+Y_t^2,
\end{equation}

where, with
$ d \tilde J^Y_t =  \gamma_+\, d\tilde{J}^+_t  + \gamma_-\,d\tilde{J}^-_t$, and $\gamma^{\pm}$ constant,


\begin{align}
Y_t &= \bar Y + \int_{-\infty}^t\,\kappa_H(t-s)  \,d\tilde J^Y_s\notag\\
&= \bar Y + \left(\kappa_H\star  \tilde J^Y\right)_t.
\end{align}

- Thus, when a trade arrives, $X$ and $\lambda$ (or $V$) jump simultaneously.


- We refer to this model as the QHO (quadratic Hawkes order flow) model.
 


### Diffusion limit of the QHO model 

- In complete analogy with the AFI case, with
\begin{align*}
\frac{d}{dt}\angl{Z,W}_t =: \rho 
&= \frac{1}{\sqrt{v_+ + v_-}}\, \frac{\gamma_+\,v_+ -\gamma_-\,v_-}{\sqrt{{\gamma_+}^2\,v_+ +
{\gamma_-}^2\,v_-}}.
\end{align*}
and
$$
\eta^2  = \frac{{\gamma_+}^2\,v_+ +
{\gamma_-}^2\,v_-}{v_+  + v_-},
$$
we obtain
<blockquote><div style="background-color:#add8e6; color:darkred; font-style: normal;  " ><h4>
The limiting model</h4>
</div>
<div style="background-color:#E8E8E8; color:#000000; font-style: normal; ">
<br>
\begin{align}
dX_t &= -\tfrac 12\,V_t\, dt + \sqrt{V_t}\,dZ_t,\notag\\
V_t &= \uv+\left(\bar Y + \int_{-\infty}^t\,\kappa(t-s)  \,\sqrt{V_s}\,dW_s \right)^2,
\label{eq:QHOlimit}
\end{align}
$\quad$ with $d\angl{Z,W}_t = \rho\,dt$, and $\kappa = \eta\,\kappa_H$.
    <p>
    <br>
</div>
</blockquote>

- This is a generalization of the QRH model with $\rho >-1$ in general.

### Related models


- The QHO model is related to the ZHawkes model of <span>[Blanc et al.]<sup id="cite_ref-BDB17" class="reference"><a href="#cite_note-BDB17">[1]</a></sup> and the AFI model of <span>[AFV]<sup id="cite_ref-AFV" class="reference"><a href="#cite_note-AFV">[10]</a></sup>.


- The diffusion limit of the QHO model is the quadratic rough Heston model of <span>[Rømer]<sup id="cite_ref-Romer" class="reference"><a href="#cite_note-Romer">[14]</a></sup>.


- When $\gamma_- = \gamma > 0$ and $\gamma_+ =  -\gamma$,
$$
\rho=-1; \qquad \eta = \gamma,
$$and we retrieve the quadratic rough Heston model of <span>[QRH]<sup id="cite_ref-QRH" class="reference"><a href="#cite_note-QRH">[9]</a></sup> and <span>[Bourgey and Gatheral]<sup id="cite_ref-BGSSR" class="reference"><a href="#cite_note-BGSSR">[4]</a></sup>. 
    - In this special case,
$ d \tilde J^Y_t = - \eta\, \left(d\tilde{J}^+_t  - d\tilde{J}^-_t \right)$.  So every time $X$ jumps up, $Y$ jumps down, and vice versa.



### The QRH simulation scheme

- Recall the dynamics of $y_{\bullet}(u)$:
$$
dy_t(u) = \kappa(u-t)\,\sqrt{V_t}\,dW_t .
$$


- The dynamics of $y$ are formally identical to those of  $\xi$ in an affine forward variance (AFV) model!  


- Thus, we may use the HQE scheme of <span>[Effiicient simulation]<sup id="cite_ref-Efficient" class="reference"><a href="#cite_note-Efficient">[7]</a></sup> to simulate $y$.


### Simulation of $Y$


- Following the derivation of the HQE algorithm closely, we have that
$$
Y_T=y_T(T) = y_t(T) + \int_t^T\,\kappa(T-s)\,\sqrt{V_s}\,dW_s.
$$


- Let $t=0$.% and the time step $\Delta = T/N$ where $N$ is the number of steps.  


- Then, with more-or-less obvious notation,
<p>
    $$
Y_n =y_n  + \sum_{k=1}^{n}\,\int^{k \Delta}_{(k-1)\Delta}\,\kappa( n\Delta - s)\,\sqrt{V_s}\,dW_s
=: \hat y_n  + u_n,
$$
    

- The $\cF_{n-1}$-adapted variable $\hat y_n$ is given by

\begin{equation*}
\hat y_n := \eefs{Y_n}{n-1} = y_n  + \sum_{k=1}^{n-1}\,\int^{k \Delta}_{(k-1)\Delta}\,\kappa( n\Delta - s)\,\sqrt{V_s}\,dW_s,
%\label{eq:yhatAnal}
\end{equation*}


- The martingale increment $u_n$ is given by
\begin{equation*}
u_{n} := \int_{(n-1) \Delta }^{n \Delta}\,\kappa(n \Delta-s)\,\sqrt{V_s}\,dW_s.
\label{eq:un} 
\end{equation*}


### The $X$-process


We also need to simulate the $n$th increment of the component of the log-stock price process $X =\log S$:
\begin{equation*}
% X_n-X_{n-1} =: 
\chi_n  := \int_{(n-1) \Delta }^{n \Delta}\,\sqrt{V_s}\,dW_s.
\label{eq:chin}
\end{equation*}


### Notation

- For convenience, define for every $i,j \in \mathbb{N}$,
<p>
\begin{align*}
\cK_{i}(\Delta) &:= \int_0^\Delta\,\kappa(s+i\Delta)\,ds\\
\cK_{i,j}(\Delta) &:= \int_0^\Delta\,\kappa(s+i\Delta) \kappa(s+j\Delta) \,ds.
\end{align*}

    
- For the gamma kernel,  
<p>
\begin{align*}
    \cK_{i}(\Delta) 
    &= 
    \frac{\nu}{\lambda^{H+1/2} \Gamma(\alpha)} 
    \left[
        \Gamma(H+1/2, i\lambda\Delta) - \Gamma(H+1/2, (i+1)\lambda\Delta)
    \right],
    \\
    \cK_{i,i}(\Delta) 
    &= 
    \frac{\nu^2}{(2\lambda)^{2H} \Gamma(\alpha)^2} 
    \left[
        \Gamma(2H, 2i\lambda\Delta) - \Gamma(2H, 2(i+1)\lambda\Delta)
    \right],
\end{align*}
<p>
where $\Gamma(s,x)=\int_x^\infty t^{s-1} e^{-t} \, dt$ is the upper incomplete gamma function. 

### Computation of $\varfs{u_n}{{n-1}}$

The conditional variance may be approximated by
\begin{align*}
\varfs{u_n}{n-1} &=\varfs{Y_n}{n-1} \\
&\approx
\left[\theta\,\hat y_n^2+  (1-\theta)\,Y_{n-1}^2 +c \right]\,K_0(\Delta)\\
&=:\bar V_n\,\cK_{0,0}(\Delta),
\end{align*}
with
$
    \theta := \frac{1}{2H+1}
$
and
\begin{equation}\label{eq:integral_resolvent}
    K_0(\Delta) := \int_0^\Delta\,K(s)\,ds,
\end{equation}
where $K$ is the resolvent kernel of $\kappa^2$.

### Simulation of $u_n$ and $\chi_n$

- As in <span>[Effiicient simulation]<sup id="cite_ref-Efficient" class="reference"><a href="#cite_note-Efficient">[7]</a></sup>, we make the hybrid approximation
$
u_n \approx \beta_{u \chi}\,\chi_n+ \varepsilon_n,
$
where $\beta_{u \chi} = \cK_0(\Delta)/\Delta$.



- We simulate $\chi_n$ and $\epsilon_n$ as
$$
\chi_n \sim \sigma_\chi\,Z_\chi; \quad \epsilon_n \sim \sigma_\epsilon\,Z_\epsilon,
$$
 where $Z_\chi $ and $Z_\epsilon$ are two independent standard normal random variables, and
 <p>
$$
\sigma_\chi^2 = \bar V_n\,\Delta;\quad \sigma_\epsilon^2 = \bar V_n\, 
\left[ \cK_{0,0}(\Delta)- \cK_0(\Delta)^2/\Delta\right],
$$


- In the large $n$ limit, $\var[u_n]$, $\var[\chi_n]$ and $\cov[u_n\,\chi_n]$ are matched.


### Simulation of $\hat y_n$

- For any $j \in \mathbb{N}$, let  $b^\star_j$ be the root mean square average of the kernel $\kappa$ over the $j$th interval
    -  This choice ensures that quadratic variations are matched exactly.
    <p><p>
\begin{equation*}
%(b^\star_{n-k+1})^2 = \sqrt{\frac 1 \Delta\, \int_{(k-1) \Delta }^{k \Delta}\,\kappa^2(n \Delta-s)\,ds} =
 b^\star_j =\sqrt{\frac 1 \Delta\, \int_{(j-1) \Delta }^{j \Delta}\,\kappa^2(s)\,ds} 
 = \sqrt{\frac 1 \Delta\, \cK_{j-1,j-1}(\Delta)}.
\end{equation*}
    
    
  -  At the $n$th step of the algorithm, we need $\hat y_n = \eefs{y_n}{n-1} $:
$$
\hat{y}_{n}  \approx y_n + \sum_{k=1}^{n-1}\,b^\star_{n-k+1}\,\chi_k.
$$

- Note the explicit path-dependence!

### Simulation of $V_n$ and $X_n$

Having now simulated $\hat y_n$ and $u_n$, we
simulate $V_n$ and $X_n = \log(S_n)$ as
<p>
\begin{equation*}\label{eq:QRH}
\begin{aligned}
V_n &= Y_n^2 + c = \left(\hat y_n +u_n\right)^2 + c,\\
X_n &= X_{n-1}-\tfrac14\,(V_n+V_{n-1})\,\Delta - \chi_n.
\end{aligned}
\end{equation*}

- At this point, we have simulated $X$ and its quadratic variation, so we can, in principle, price all SPX derivatives, including variance swaps.


### Simulation of VIX

- Let $\delta = 30/365$ (30 days). 


- Given $y_T(u)$, $u>T$, we can show that, under quadratic rough Heston,
<p>
\begin{align}\tag{3}
\VIX^2_T 
%&= 
%\frac 1 \delta\,\int_T^{T+\delta}\,\xi_T(u)\,du
%\nonumber\\
%&= 
%\frac 1 \delta\,\int_T^{T+\delta}\,
%\left\{ y_T(u)^2 + c + \int_{T}^u\,K(u-r)\, [y_T(r)^2 + c] \, dr\right\}\,du
%\nonumber\\
%&= 
%\frac 1 \delta\,\int_T^{T+\delta}\,
%(y_T(u)^2 + c) \,\left\{ 1 + \int_{u}^{T+\delta}\,K(r-u)\,dr\right\}\,du
%\nonumber\\
%&=  \frac 1 \delta\,\int_T^{T+\delta}\,\left[y_T(u)^2+c\right]\,\left(1 + K_0(T+\delta-u)\right)\,du\nonumber\\
&=  
\frac 1 \delta\,\int_0^{\delta}\,(y_T(T+u)^2 + c)
\,\left\{1 + K_0(\delta-u)\right\}\,du,
\end{align}
where $K_0(\tau) =  \int_{0}^{\tau}\,K(s)\,ds$.


### Simulation of the $y_T(u)$

For any $u>T$,
\begin{align}
y_T(u) &= y(u) + \int_{0}^T\,\kappa(u-s)\,\sqrt{V_s}\,dW_s\\
%&=& y(u) + \sum_{k=1}^{n}\,\int^{k \Delta}_{(k-1)\Delta}\,\kappa(u-s)\,\sqrt{V_s}\,dW_s\\
&\approx y(u) + \sum_{k=1}^{m}\,a_k(u)\,\int^{k \Delta}_{(k-1)\Delta}\,\sqrt{V_s}\,\,dW_s\\
&= y(u) + \sum_{k=1}^{m}\,a_k(u)\,\chi_k,%\\
\end{align}
<p>
where the $a_k(u)$ are chosen 
so as to match the variances of the $y_T(u)$. 


### Final VIX simulation recipe

- Plugging into (3) gives<p>
\begin{align*}
\VIX^2_T 
&=  \frac 1 \delta\,\int_0^{\delta}\,(y_T(T+u)^2 + c)\,\left(1 + K_0(\delta-u)\right)\,du\nonumber\\
% &=  \frac 1 \delta\,\int_0^{\delta}\,\left( (y_t(T+u) + c) 
% + \sum_{k=1}^{n}\,a_k(T+u)\,\chi_k  \right)^2\,\left[1 + K_0(\delta-u)\right]\,du,
&=  \frac 1 \delta\,\int_0^{\delta}\,
\Big[ \Big(y(T+u) + \sum_{k=1}^{n}\,a_k(T+u)\,\chi_k \Big)^2 + c \Big]\notag\\
&\qquad\qquad\qquad\qquad\qquad\qquad \times \Big[ 1 + K_0(\delta-u) \Big] \, du.
%\label{eq:VIX2a}    
\end{align*}
<p>

- A numerical integral that may be easily computed.
    - For example, using the trapezoidal rule or a Gauss-Legendre quadrature.

### Path-dependence of the QR Heston model

- From (3), changes in the forward variance curve are directly expressible in terms of changes in the underlying.  
    - The QR Heston model is purely path-dependent.


- Explicitly, using $
dy_t(u) = \kappa(u-t)\,\sqrt{V_t}\,dW_t $, we may write
\begin{align*}
y_t(u) &=\bar Y + \int_{-\infty}^t\,\kappa(u-r)\,\sqrt{V_r}\,dW_r=\bar Y - \int_{-\infty}^t\,\kappa(u-r)\,\frac{dS_r}{S_r}.
\end{align*}
    - The forward volatility curve is thus a weighted average of historical stock returns.


- The fair values of SPX options, VIX options, and the SSR are all functionals of the forward volatility curve.

### The skew-stickiness ratio (SSR)

- Recall that  the SSR is given by
<p>
$$
\mathcal{R}_t(T) = \frac{\beta_t(T)}{\cS_t(T)} 
$$
where
$\beta_t(T)$ is the regression coefficient  in 
<p>
    $$
\delta \sigma_t(T) = \beta_t(T)\,\delta X_t + \text{noise},
%\label{eq:betaDef}
$$
where $\sigma_t(T)$ is ATM vol. and $\cS_t(T)$ is the ATM skew.
   
    
- In practice whole smiles are tied to the ATM vol.  So this is a recipe for real-time updating of the whole vol. surface.

### Estimation of the regression coefficient $\beta_t(T)$ under QRH

- Recall that in martingale form, the QR Heston model reads:
$$
dy_t(u) = \kappa(u-t)\,\sqrt{V_t}\,dW_t.
$$


- We apply the finite difference methodology of <span>[Bourgey et al.]<sup id="cite_ref-SSRFD" class="reference"><a href="#cite_note-SSRFD">[2]</a></sup>.


- Thus, the regression coefficient is estimated as
<p>
 $$ \beta_{t}(T)
    \approx \frac 1 h\,\left\{
    \sigma_{t}\left(T,y^h_{t}(\cdot)\right )
    -
    \sigma_{t}\left( T, y^{}_{t}(\cdot)\right)
    \right\},
    \label{eq:ssr-finite-difference}
    $$
where $h$ is a small parameter, and for $t \leq u \leq T$,
\begin{equation}\label{eq:y_blip}
    y^h_{t}(u) := y_t(u) - h\,\kappa(u-t).
\end{equation}

### Blipped forward volatility curves


- Experiment shows that scaling the blip by time to expiration works well.


- Specifically, we choose $h=-0.01 \sqrt{T}$.


- $h<0$ ensures a positive blipped forward variance curve.


Note that Figure 3 in the paper has more expiries, 300,000 paths, and 300 time steps. We have reduced the computational burden here for illustration.

```python
paths = 1 * 10**5
steps = 100
h_ssr = lambda t: -1e-2 * np.sqrt(t)
expiries = 2 ** (np.arange(-13, 1) / 2)
n_expiries = len(expiries)
```

```python
qrh_sim_ssr = qrh.simulate(
    paths=paths,
    steps=steps,
    spx_expiries=expiries,
    output="all",
    h_ssr=h_ssr,
)
```

### Compute base and blipped implied volatilities

```python
S_matrix = np.array([np.exp(qrh_sim_ssr["spx"][T]["X"]) for T in expiries])
S_h_matrix = np.array([np.exp(qrh_sim_ssr["spx"][T]["X_h"]) for T in expiries])
```

### Illustrate for one expiration


```python
k = np.arange(-0.4, 0.2, 0.001)
idx = 8
impvols_mc = black_otm_impvol_mc(S_matrix[idx, :], k=k, T=expiries[idx])
impvols_h_mc = black_otm_impvol_mc(S_h_matrix[idx, :], k=k, T=expiries[idx])
```

```python
fig, ax = plt.subplots()
ax.plot(k, impvols_mc, color="blue", linestyle="-", label="Base Volatility")
ax.plot(k, impvols_h_mc, color="red", linestyle="--", label="Blipped Volatility")
ax.set_xlabel("Log-strike k")
ax.set_ylabel("ATM volatility")
ax.legend()
plt.show()
```

Figure 9: Base smile in blue; blipped smile in red.

### Computation of the regression coefficient $\beta$


```python
atm_impvols_mc = np.array(
    [
        black_otm_impvol_mc(S_matrix[idx, :], k=0.0, T=expiries[idx])
        for idx in range(n_expiries)
    ]
).flatten()
atm_impvols_h_mc = np.array(
    [
        black_otm_impvol_mc(S_h_matrix[idx, :], k=0.0, T=expiries[idx])
        for idx in range(n_expiries)
    ]
).flatten()
beta = (atm_impvols_h_mc - atm_impvols_mc) / h_ssr(expiries)
```

### Log-log plot of $\beta_t(T)$


```python
fig, ax = plt.subplots()
ax.plot(np.log(expiries), np.log(np.abs(beta)), ".b")
ax.set_xlabel("Maturity T")
ax.set_ylabel(r"$\beta(T)$")
plt.show()
```

### Estimation of the ATM skew $\cS_t(T)$

- Recall that the ATM skew
<p>
$$\cS_t(T) = \left. \frac{\partial \sigma_{\BS}(k,T)}{\partial k}\right|_{k=0}.$$
    
    
    
- We estimate the ATM skew using a trick from <span>[Bourgey, Friz et al.]<sup id="cite_ref-LocalRough" class="reference"><a href="#cite_note-LocalRough">[3]</a></sup>: for any $k \in \R$,
<p>
    $$
    \frac{\partial \sigma_{\BS}(k,T)}{\partial k}
    =
    \left. \frac{\Phi\left(-\frac{k}{\sqrt{w}} - \frac{\sqrt{w}}{2}\right) - \mathbb{P}\left(X_T \geq k\right)}{\sqrt{T} \phi\left(-\frac{k}{\sqrt{w}} - \frac{\sqrt{w}}{2}\right)} \right|_{w=T \sigma_{\BS}^2(k,T)}
$$
    <p>
where $X_T = \log(\frac{S_T}{S_0})$ and $\phi$ is the p.d.f. of a standard normal distribution.

### Code to compute the ATM skew $\cS_t(T)$

```python
w = expiries * atm_impvols_mc**2
atm_digit = (S_matrix >= S_matrix.mean(axis=1, keepdims=True)).mean(axis=1)
atm_skew = norm.cdf(-0.5 * w**0.5) - atm_digit
atm_skew /= expiries**0.5 * norm.pdf(-0.5 * w**0.5)
```

### Log-log plot of $\mathcal{S}_t(T)$


```python
fig, ax = plt.subplots()
ax.plot(np.log(expiries), np.log(np.abs(atm_skew)), ".b")
ax.set_xlabel("Maturity T")
ax.set_ylabel("ATM Skew")
plt.show()
```

### Estimation of the SSR

-  Finally, the SSR is estimated as
\begin{equation}
    \cR_{t}(T) \approx \frac{\beta_t(T)}{\cS_t(T)}.\tag{4}
\end{equation}



- The estimator (4) only requires only 
the ATM implied volatilities for two different forward volatility curves $y_t(u)$, and one estimate of the ATM skew.


```python
ssr = beta / atm_skew
```

### Why does the QR Heston model fit so well?

- In view of its extreme parsimony, it may at first seem surprising that the QR Heston model can generate such impressive fits.


- By taking a historical perspective and relating the QR Heston model to earlier models,  we may begin to understand its strong fitting performance.

### QR Heston from Quadratic Hawkes

- The QR Heston model in (1) is a special case of the quadratic rough Heston model of <span>[Dandapani et al.]<sup id="cite_ref-DJR" class="reference"><a href="#cite_note-DJR">[5]</a></sup>.


- This model, in turn, emerges as the continuous-time limit of a non-Markovian extension of the quadratic Hawkes model originally proposed in <span>[Blanc et al.]<sup id="cite_ref-BDB17" class="reference"><a href="#cite_note-BDB17">[4]</a></sup>.


- It, in turn, can be viewed as a limit of the QARCH (quadratic ARCH) model.  
    - Calibration of QARCH to 5-minute intra-day US stock yielded very impressive results.  
    
    
- The QUARCH model has been used in practice for a long time, in particular by Bouchaud at CFM.

    
- Thus the QR Heston model can be seen as a non-Markovian extension of a long-standing successful statistical model of financial markets.

### Markovian QR Heston

- The QR Heston model can also be seen as the non-Markovian extension of various realistic models in the classical
stochastic volatility literature.


- Consider the QR Heston model in the Markovian case ($H=1/2$) with the gamma kernel (2):<p>
$$\kappa(\tau) =\nu\,e^{-\lambda\,\tau},$$ and  $\uv=0$.


- In this case, $\sigma_t = Y_t$ is the instantaneous volatility, and $Y_t = \bar Y +\int_{-\infty}^t \kappa(t-s)\,\sqrt{V_s}\,dW_s$ may be rewritten as
<p>
$$
d\sigma_t = -\lambda\,(\sigma_t - \bar{Y}) \,dt + \nu \,\sigma_t\,dW_t, \tag{5}
\label{eq:IG}
$$

### QR Heston and the Inverse Gamma model

- (5) are the dynamics of volatility in the Inverse Gamma model of <span>[Langrené et al.]<sup id="cite_ref-LLZ" class="reference"><a href="#cite_note-LLZ">[13]</a></sup>. 


- As noted in that paper, the Inverse Gamma model is a special case of various other models that have desirable properties consistent with empirical studies of stocks and their volatilities. 


-  For example, the Inverse Gamma model has an inverse gamma stable distribution, which was shown to be a good fit to the unconditional distribution of volatility by Bouchaud and Potters and Artur Sepp amongst others.
    - In fact, this observation was one of the reasons that Sepp chose his particular model dynamics.


### QR Heston and the Double Lognormal model

- In terms of $V_t=\sigma_t^2 = Y_t^2$, (5) may be written in the form
<p>
\begin{equation*}\label{eq:DL}
\begin{aligned}
dV_t &=-\hat \lambda\,\left(V_t-V_t^{\prime}\right)\,dt +2\, \nu \,V_t\,dW_t,\\
dV^\prime_t &= - \lambda\,\left(V^\prime_t-\bar{V}^\prime\right)\,dt + \nu \,V'_t\,dW_t,
\end{aligned} 
\end{equation*}

where

\begin{align*}
\hat \lambda =2\, \lambda-\nu^2; \qquad V^\prime_t = \frac{2\lambda}{\hat \lambda}\,\bar{Y} \, \sigma_t;
\qquad \bar{V}^\prime = \frac{2\lambda}{\hat \lambda}\,\bar{Y}^2.
\end{align*}

- This is a special case of the Double Lognormal (DL) model of <span>[Bachelier 2008]<sup id="cite_ref-DL" class="reference"><a href="#cite_note-DL">[8]</a></sup>
  - DL generates positively sloping VIX smiles
  - DL  parameters are reasonable stable.

### Lifting the Double Lognormal model}

- Viewing the gamma kernel as an infinite sum of exponential kernels with different characteristic timescales in the spirit of Abi Jaber, the QR Heston model may be seen as an infinite sum of Double Lognormal models.


- Accordingly, the QR Heston model inherits various properties of the Double Lognormal model such as positively sloped VIX smiles.

### Summary

- The Quadratic Rough Heston (QR) Heston model can successfully capture the joint behavior of SPX and VIX implied volatility smiles, whilst producing realistic values for the SSR. 
    - Thus statically and dynamically reasonable.





- It may be derived as the limit of a quadratic Hawkes proces describing order flow.


- The QR Heston model with the gamma kernel is:
    - Extremely parsimonious with only four parameters
    -Purely path-dependent


- The QRH scheme is efficient and easy to implement. 
    - The QR Heston model is thus a viable candidate for practical trading applications.
